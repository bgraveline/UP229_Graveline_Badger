```python
import contextily as ctx
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(25,25))
color_mapping = {"A": "green", "B": "blue", "C":"yellow","D":"red"}
gdf.to_crs('EPSG:3857').plot('holc_grade',color=gdf["holc_grade"].map(color_mapping),legend = True, ax=ax, alpha= 0.7)
ctx.add_basemap(ax, zoom = 12, source = ctx.providers.Stamen.TonerLite)

ax.set_xticks([])
ax.set_yticks([])

ax.set(title='HOLC Redlining') 
```
